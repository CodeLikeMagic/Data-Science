```python
import math

def area(r):
    return math.pi * (r**2)


radii = [1, 2, 3, 4, 5]
areas = []

for r in radii:
  a = area(r)
  areas.append(a)
```

```python
map(area, radii)

list(map(area, radii))
```

```python
import statistics
data = [3, 6, 8 ,6 , 2] 
avg = statistics.mean(data)
print(float(avg))

filter(lambda x: x>avg, data)

list(filter(lambda x: x>avg, data))
```

```python
from functools import reduce

data = [1,2,3,4,5]
multiplier = lambda x,y: x*y
reduce(multiplier, data)

'''
val1 = 1*2
val2 = val1*3
val3 = val2*4
val4 = val3*5
return val4
'''

```
